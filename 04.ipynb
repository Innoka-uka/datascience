```python
import matplotlib.pyplot as plt
import numpy as np

# 
x = np.linspace(0, 10, 100)  # 010100
y = np.sin(x)                 # y  x 

```

```python
# 1. 
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='b')
plt.title('Line Plot of Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid()
plt.show()

# 2. 
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='orange')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# 3. 
x_scatter = np.random.rand(50)  # 50
y_scatter = np.random.rand(50)

plt.figure(figsize=(10, 6))
plt.scatter(x_scatter, y_scatter, color='green', alpha=0.5)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 
np.random.seed(0)
data = np.random.randn(100)

# 1.  (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='skyblue')
plt.title('Histogram with KDE')
plt.show()

# 2.  (Categorical Scatterplot)
#  tips 
tips = sns.load_dataset("tips")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips, palette='deep')
plt.title('Categorical Scatterplot')
plt.show()

# 3.  (Heatmap)
# 
corr_matrix = np.corrcoef(np.random.randn(10, 200)) 
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```
